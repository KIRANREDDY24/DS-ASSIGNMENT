```python
import csv
with open("assignment_data.xlsx - Subscription Bought.csv", newline='') as csvfile:
    spamreader = csv.reader(csvfile)
    with open('greater499.csv', 'w', newline='') as csvfile1:
        spamwriter = csv.writer(csvfile1)
        next(spamreader)
        for row in spamreader:
            if int(row[5]) >= 499:
                spamwriter.writerow(row)
```
